# 1/ Import Packages

```python
import numpy as np
import pandas as pd
```

# 2/ Import Data

```python
df = pd.read_csv('OrderDetails.csv')
```

# 3/ Covert OrderDate to datetime
- extract Year and Month

```python
df['OrderDate'] = df['OrderDate'].astype('datetime64[ns]')
df['Ord_Year'] = pd.DatetimeIndex(df['OrderDate']).year
df['Ord_Month'] = pd.DatetimeIndex(df['OrderDate']).month
```

## Check Data

```python
df.head()
```

```python
df.info()
```

# 4/ Create Summary

```python
summary = df.groupby(['Ord_Year', 'Ord_Month'], as_index = False)['Quantity'].sum()
summary
```

# 5/ Export Summary

```python
summary.to_csv('python_summary.csv', index = False)
```
